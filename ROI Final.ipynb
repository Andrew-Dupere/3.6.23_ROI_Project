```python
import pandas as pd
from tabulate import tabulate
from IPython.display import clear_output


class ROI():    
    def __init__(self):
        self.expenses = {}
        self.income = {}
        self.investment = {}
    
    def __get_investment(self):
        print("""
        Please enter your initial investments in the format:
        "Investment Item: Dollar Amount"
        """ )
        try:
            Iprompt = input()
            self.investment[Iprompt.split(':')[0]] = int((Iprompt.split(':')[1]).strip().strip('$'))
        except:
            pass
        self.run()
    
    def __get_expenses(self):
        print("""
        Please enter your expenses in the format:
        "Expense Item: Dollar Amount"
        """ )
        try:
            Eprompt = input()
            self.expenses[Eprompt.split(':')[0]] = int((Eprompt.split(':')[1]).strip().strip('$'))
        except:
            pass
        self.run()
    
    def __get_income(self):
        print("""
        Please enter your rental income in the format:
        "Income Item: Dollar Amount"
        """ )
        try:
            Incprompt = input()
            self.income[Incprompt.split(':')[0]] = int((Incprompt.split(':')[1]).strip().strip('$'))
        except:
            pass
        self.run()
    
    def __display_info(self):
        
        investmentdf = tabulate(pd.DataFrame((key,val) for key, val in self.investment.items()),headers='',tablefmt='github')    
        expensesdf = tabulate((pd.DataFrame((key,val) for key, val in self.expenses.items())),headers='',tablefmt='github')
        incomedf = tabulate((pd.DataFrame((key,val) for key, val in self.income.items())),headers='',tablefmt='github')
        
        ROI = (((sum(self.income.values()) - sum(self.expenses.values())) * 12 )/ sum(self.investment.values()))*100
                      
        print(f"Investment ")
        print(f"{investmentdf}")
        print(f"Expenses ")
        print(f"{expensesdf}")
        print(f"Income ")
        print(f"{incomedf}")
        print(f"ROI:{ROI:.2f}%")
        
        self.run()
    
    def __delete_item(self):
        
        print("""
        Press 1 to delete an investment item
        Press 2 to delete an income item
        Press 3 to delete an expense item        
        """)
             
        dprompt = input()
            
        if dprompt == '1':
            print('Investment Items:')        
            for key in self.investment.keys():           
                print(key)
            removal = input('Which item would you like to remove?')
            del self.investment[removal]
            print(f'{removal} has been removed from the cart')

        if dprompt == '2':                 
            print('Income Items:')
            for key in self.income.keys():
                print(key)
            removal = input('Which item would you like to remove?')
            del self.income[removal]
            print(f'{removal} has been removed')

        if dprompt == '3':
            print('Expenses Items:')
            for key in self.expenses.keys():
                print(key)
            removal = input('Which item would you like to remove?')
            del self.expenses[removal]
            print(f'{removal} has been removed')

        else:
            self.run()

    def run(self):
        
        print("""
        ROI Calculator
                
        Press 1 to input investment information
        Press 2 to input income information
        Press 3 to input expense information
        press 4 to view financial information
        press 5 to edit financial information
        press 6 to exit
        """)
        prompt = input()
        
        while prompt != '6':

            
            if prompt == '1':
                clear_output()
                self.__get_investment()
            elif prompt == '2':
                clear_output()
                self.__get_income()
            elif prompt == '3':
                clear_output()
                self.__get_expenses()                
            elif prompt == '4':
                clear_output()
                self.__display_info()
            elif prompt == '5':
                clear_output()
                self.__delete_item()

            else:
                print('Invalid Input')
                self.run()
                
        if prompt == '6':
            print("GoodBye")
            
            
```
